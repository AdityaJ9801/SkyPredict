```python
import pandas as pd

# Read data from data.csv file
data = pd.read_csv('weatherAUS.csv')

data
```

```python
# Select the required columns
selected_data = data[['Date', 'Location', 'MinTemp', 'MaxTemp', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'RainToday', 'RainTomorrow']]

selected_data
```

```python
# Calculate the averages
selected_data['Temp'] = selected_data[['MinTemp', 'MaxTemp']].mean(axis=1)
selected_data['WindSpeed'] = selected_data[['WindSpeed9am', 'WindSpeed3pm']].mean(axis=1)
selected_data['Humidity'] = selected_data[['Humidity9am', 'Humidity3pm']].mean(axis=1)

# Select final columns to save in new CSV
final_data = selected_data[['Date', 'Location', 'Temp', 'WindSpeed', 'Humidity', 'RainToday', 'RainTomorrow']]

final_data
```

```python
# Save the result to a new CSV file
final_data.to_csv('processed_weather_data.csv', index=False)

print("Averaged data saved to 'processed_weather_data.csv'")
```
